```python
import pandas as pd
import random
import os
```

```python

df = pd.read_excel('./문제와정답.xlsx')
list_question = list(range(len(df)))
```

```python
while len(list_question) >=1 :
    i = random.randrange(len(list_question))
    question = df['설명'][i].replace('\n', ' ')
    question = df['설명'][i].replace('/', '\n')

    print('-'*150)
    print('문제 :', question)
    print('\n')
    print('정답은?')
    print('\n')

    ans = input()
    print('입력 :',ans)

    answer = df['정답'][i]
    print('정답 :', answer)

    print('-'*150)
    list_question.pop(i)
```
