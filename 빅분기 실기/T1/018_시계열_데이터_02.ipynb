**주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하라 (소수점 둘째자리까지 출력, 반올림)**

```python
# 라이브러리 불러오기
import pandas as pd

# 데이터 불러오기 (datatime컬럼 지정)
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv", parse_dates=['Date'])

df.head()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```

```python
df = df[(df['year'] == 2022) & (df['month'] == 5)]
```

```python
# 주말 평일 구분
df['dayofweek'] = df['Date'].dt.dayofweek
df['weekend'] = df['dayofweek'].apply(lambda x: x>=5)
```

```python
weekend = df[df['weekend'] == True]['Sales'].mean()
weekday = df[df['weekend'] == False]['Sales'].mean()
```

```python
weekend, weekday
```

```python
print(round(weekend - weekday, 2))
```
