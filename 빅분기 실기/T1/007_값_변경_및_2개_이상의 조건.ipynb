**'f4'컬럼의 값이 'ESFJ'인 데이터를 'ISFJ'로 대체하고, 'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age'컬럼의 최대값을 출력**

```python
# 라이브러리 호출
import pandas as pd
import numpy as np

# 데이터 불러오기
df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df
```

```python
# f4가 ESFJ인 값을 ISFJ로 대체
df['f4'] = df['f4'].replace('ESFJ', 'ISFJ')
```

```python
# 2개의 조건에 맞는 값
df[(df['f4'] == 'ISFJ') & (df['city'] == '경기')]
```

```python
# 2개의 조건에 맞는 값중 age컬럼의 최대값
df[(df['f4'] == 'ISFJ') & (df['city'] == '경기')]['age'].max()
```
