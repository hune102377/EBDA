**basic1 데이터와 basic3 데이터를 'f4'값을 기준으로 병합하고, 병합한 데이터에서 r2결측치를 제거한다음, 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합을 구하라**

```python
# 라이브러리 및 데이터 로드
import pandas as pd
b1 = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")
b3 = pd.read_csv("../input/bigdatacertificationkr/basic3.csv")
```

```python
df = pd.merge(left = b1, right = b3, how = 'left', on = 'f4')
df
```

```python
df = df[~df['r2'].isnull()]
```

```python
df = df.iloc[:20]
print(df['f2'].sum())
```
