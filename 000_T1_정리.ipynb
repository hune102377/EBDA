### **정의 관련**

- **IQR**  
: 3분위(0.75) 값에서 1분위(0.25) 값을 뺀 범위
  
- **이상치**  
: (3분위 + 1.5 * IQR)보다 크고 (1분위 - 1.5 * IQR)보다 작은 값들   

- **왜도(s)**  
: 실수 값 확률 변수의 확률 분포 비대칭성을 나타내는 지표  
평균과 중앙값이 같으면 왜도는 0이 된다 

- **첨도(k)**  
: 확률분포의 꼬리가 두꺼운 정도,  
k=3인 경우 정규분포  

- **~**  
: '반대'의 의미 (**6번 문제 다시 풀어보기**)

- **&**  
: '그리고'의 의미 (**7번 문제 다시 풀어보기**)

- **상관관계**  
: '상관관계의 크기 유무는 절대값으로 계산해야한다  
ex) 0.4보다 -0.7이 상관관계가 크다

- **sigmoid**  
: 1 / (1 + np.exp(-x))

- **상관관계**  
: np.maximum(0, x)

### **기본 함수 관련**

- **max()**  
: 최대값

- **min()**  
: 최소값

- **mean()** 
: 평균값

- **median()** 
: 중앙값

- **mode()**  
: 최빈값

- **sum()**  
: 합계

- **count()**
: 데이터 수(결측값 제외됨)

- **cumsum()**  
: 컬럼의 누적합

- **fillna()**  
: 결측치 처리  
**method 파라미터를 bfill로 설정하면 빈 값의 경우 뒤에 나오는 값으로 채움)** 

- **round(n, a)**  
: n 값의 a번째 자리까지 출력

- **groupby()**  
: 특정 기준으로 데이터프레임을 묶어주는 함수(**6번 문제 다시 풀어보기**)

- **replace(a, b)**  
: 컬럼의 a를 b로 변환

- **drop()**  
: 인덱스 삭제, axis = 1로 파라미터를 두면 컬럼을 삭제

- **unique()**  
: 컬럼의 고유값 확인

- **value_counts()**  
: 컬럼의 각 고유값의 수 카운트

- **sort_values()**  
: ascending = False(내림차순, 큰것에서 작은 순)

- **isnull()**  
: 결측치 확인

- **map()**  
: 함수 전체 적용 (**3번 문제 다시 풀어보기**)

- **quantile(n)**  
: 0.n 분위수 확인

- **var()**  
: 컬럼의 분산 확인

- **skew()**  
: 컬럼의 왜도값 확인

- **kurt()**  
: 컬럼의 첨도값 확인

- **std()**  
: 컬럼의 표준편차 확인

- **drop_duplicates()**  
: subset 파라미터로 컬럼 지정 가능

- **qcut(df['target'], q=n**  
: 특정 컬럼을 n개의 그룹으로 나눈다

- **merge(left= left_df, right = right_df, howe = 'left', on = a)**  
: a컬럼을 기준으로 left_df에 right_df 을 병합

- **qcut(df['target'], q=n**  
: 특정 컬럼을 n개의 그룹으로 나눈다

- **str.cotains('text')**  
: 특정 컬럼에 text라는 단어가 포함되어있으면 True 값을 반환

- **apply()**  
: 컬럼에 특정함수 적용
